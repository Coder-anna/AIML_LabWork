# Naive Bayes Classifier 

Using Naive Bayes to predict the type of insurance person has in the DoctorAUS dataset

```python
from pydataset import data 
import pandas as pd 
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report 
```

```python
pip install pydataset
```

```python
df= data("DoctorAUS")
X= df[['age', 'income', 'sex', 'illness', 'actdays', 'hscore', 'doctorco', 'nondocco', 'hospadmi', 'hospdays', 'medecine', 'prescrib']]
y= df['insurance']
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=3, random_state=0)
```

```python
clf= GaussianNB()
clf.fit(X_train, y_train)
```

```python
y_pred= clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

We can se that our overall numbers are not that great. This means that the current algo is probably not the best choice for classification
