# define x,y

```python
from sklearn.datasets import load_iris
iris= load_iris()
x= iris.data
y= iris.target
```

# preprocessing
# train test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size= 0.25, stratify=y, random_state= 40)
```

# sacling

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

# building the model

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
model= DecisionTreeClassifier(max_depth=4)
model.fit(x_train,y_train)
```

# prediction and evaluation

```python
#prediction
y_pred= model.predict(x_test)
labels_order =[0,1,2]
from sklearn.metrics import f1_score
f1= f1_score(y_true= y_test, y_pred= y_pred, labels= labels_order, average= "weighted")
print(f"f1: {f1}")
```

```python
#accuracy
from sklearn.metrics import accuracy_score
acc= accuracy_score(y_true= y_test, y_pred= y_pred)
print(f"Accuracy: {acc}")
```

# plot the graph

# dark boxes are more pure than lighter ones as they belong to the same class

```python
plt.figure("Decision Tree", figsize= [13,6])
plot_tree(model, fontsize= 10, filled= True, feature_names= iris.feature_names)
plt.tight_layout()
plt.show()
```
